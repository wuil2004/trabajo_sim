## Representacion grafica de datos.

Si a la funcion de trazado se le da una matriz de datos, la usara como coordenadas en el eje vertical, y utilizara el indice de cada punto de datos en el array como la coordenada horizontal

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

Tambien se puede proporcionar dos matrices una para el eje horizontal, y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
```

pueden modificarse las lo0ngitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) #[xmin, xmax, ymin, y max]
plt.show()
```

se sigue el mismo procedimiento para pintar una funcion matematica

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

tambien puede modificarse el estilo de la grafica para que contenga mas informacion. 

```python
plt.plot(x, y)
plt.title("Square funcion")
plt.xlabel("chido")
plt.ylabel("y = **2")
plt.grid(True)
plt.show()
```

pueden superponerse graficas y cambiar el estilo de las funciones

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square funcion")
plt.xlabel("chido")
plt.ylabel("y = **2")
plt.grid(True)

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
# separando en diferentes lineas las funciones
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square funcion")
plt.xlabel("chido")
plt.ylabel("y = **2")
plt.grid(True)

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square funcion")
plt.xlabel("chido")
plt.ylabel("y = **2")
plt.grid(True)

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x+1")
plt.legend(loc = "best") # la situa en la mejor localizacion
plt.show()
```

Tambien puede crearce dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denomina sublots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2 ,1 ) # 1 Rows 2 Columns, lst Subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Rows 2  Colums, 2nd Suplots
plt.plot(x, y2, 'g')

plt.show()
```

para que las grificas no queden tan ajustadas se puede hacer la figura mas grande

```python
plt.figure(figsize = (14,6) )

plt.subplot(1, 2 ,1 ) # 1 Rows 2 Columns, lst Subplot
plt.plot(x, y, 'b--')
plt.title("Square funcion", fontsize = 14)
plt.xlabel("chido", fontsize = 14)
plt.ylabel("y = **2", fontsize = 14)
plt.grid(True)


plt.subplot(1, 2, 2) # 1 Rows 2  Colums, 2nd Suplots
plt.plot(x, y2, 'g')
plt.title("Square funcion", fontsize = 14)
plt.xlabel("chido2", fontsize = 14)
plt.ylabel("y = **2", fontsize = 14)
plt.grid(True)
```

## Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random  import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c= 'blue') 
plt.show
```

##  Histogramas 

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth= 0.8)

plt.subplot(212)
plt.hist(data, bins= [1, 1.5, 2, 2.5,3], rwidth= 0.05)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
